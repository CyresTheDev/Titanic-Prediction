```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
df = pd.read_csv('../input/datafram/final-df.csv')
test = pd.read_csv('../input/finaltestdf/final-test-df.csv') 
```

```python
df.head()
```

```python
df.head()
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
df.head(2)
```

```python
X = df.drop('Survived', axis=1)
y = df['Survived']
```

```python
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(X, y, random_state=42)
```

```python

model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(8, activation=tf.keras.activations.relu), # hidden layer 1, ReLU activation
    tf.keras.layers.Dense(4, activation=tf.keras.activations.relu), # hidden layer 2, ReLU activation
    tf.keras.layers.Dense(1, activation=tf.keras.activations.sigmoid)
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(lr=0.1),
    loss=tf.keras.losses.BinaryCrossentropy(),
    metrics=['accuracy']
)

model.fit(train_x, train_y, epochs=200, verbose=1)
```

```python
model.evaluate(train_x, train_y)
```

```python
xx = test
xx = xx.drop(['Unnamed: 0', 'PassengerId'], axis=1)
```

```python
xx = StandardScaler().fit_transform(xx)
```

```python
results = model.predict(xx)
results[results > 0.5] = 1
results[results < 0.5] = 0
results = tf.squeeze(results)
```

```python
results = np.array(results)
```

```python
import csv

ids = test['PassengerId']
filename = 'Tensorflow-Titanic.csv'

csvArray = []

for i in range(0, len(results)):
    csvArray.append([ids[i], results[i]])

with open(filename, 'w') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(['PassengerId', 'Survived'])
    csvwriter.writerows(csvArray)
```
